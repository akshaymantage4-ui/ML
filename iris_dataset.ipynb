```python
for feature in df.select_dtypes(include="number"):
  sns.lineplot(x=df[feature],y="species",data=df)
  plt.title(f"Relationship between {feature} and Species")
  plt.xlabel(f"{feature}")
  plt.ylabel("Species")
  plt.show()
```

```python
sns.countplot(df["species"])
plt.show()
```

```python
corr=df.iloc[:,:-1].corr()
plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation between featuers")
```

```python
for x in df.select_dtypes(include="float64"):
  sns.displot(df[x])
  plt.title(f"NOrmal distribution of {x}")
  plt.show()
```
